```python
import matplotlib.pyplot as plt
import numpy as np

# Define x values
x = np.linspace(0, 30, 400)

# Constraint 1: 3x + 5y <= 50 → y <= (50 - 3x)/5
y1 = (50 - 3*x)/5

# Constraint 2: 2x + 4y <= 80 → y <= (80 - 2x)/4 = (40 - x)/2
y2 = (40 - x)/2

# Make plot
plt.figure(figsize=(8,6))

# Plot constraint lines
plt.plot(x, y1, label=r'$3x + 5y \leq 50$', color='blue')
plt.plot(x, y2, label=r'$2x + 4y \leq 80$', color='green')

# Fill feasible region
y = np.minimum(y1, y2)
y = np.maximum(y, 0)  # Ensure y >= 0
plt.fill_between(x, 0, y, where=(y>=0), color='lightgrey', alpha=0.5)

# Label axes and add legend
plt.xlim((0, 30))
plt.ylim((0, 25))
plt.xlabel('x (Product A)')
plt.ylabel('y (Product B)')
plt.title('Feasible Region for LP Problem')
plt.legend()
plt.grid(True)

# Mark feasible corner points
plt.plot(0, 10, 'ro')
plt.text(0, 10.5, '(0,10)', fontsize=9)
plt.plot(16.67, 0, 'ro')
plt.text(16.67, 1, '(16.67,0)', fontsize=9)

plt.show()
```
